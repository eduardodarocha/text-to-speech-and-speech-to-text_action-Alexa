```python
from gtts import gTTS
text_to_say = "How are you doing?. What's up? "
language = "en"
gtts_object = gTTS(text = text_to_say,
                   lang = language,
                   slow = False)
gtts_object.save("/content/gtts.wav")
```

```python
from IPython.display import Audio
Audio("/content/gtts.wav")
```

```python
french_text = "Je vais au supermarché aujourd'hui"
french_language = "fr"
french_gtts_object = gTTS(text = french_text, lang = french_language, slow = False)
french_gtts_object.save("/content/french.wav")
Audio("/content/french.wav")
```

```python
portuguese_text = "A Língua Portuguesa é falada tanto em Portugal, quanto no Brasil e em outros diversos países ao redor do planeta. Assim sendo, falar este idioma pode abrir uma série de oportunidades pessoais e até mesmo profissionais."
portuguese_language = "pt"
portuguese_gtts_object = gTTS(text = portuguese_text, lang = portuguese_language, tld = 'pt', slow = False)
portuguese_gtts_object.save("/content/portuguese.wav")
Audio("/content/portuguese.wav")
```

```python
portuguese_br_gtts_object = gTTS(text = portuguese_text, lang = portuguese_language, tld = 'com.br', slow = False)
portuguese_br_gtts_object.save("/content/portuguese_br.mp3")
Audio("/content/portuguese_br.mp3")
```
